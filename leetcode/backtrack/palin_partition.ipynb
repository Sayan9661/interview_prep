```python
#palindrome function

def isPalindrome(str):
    n = int((len(str)-1)/2)
    print(n)
    while(n >= 0):
        if(str[n] != str[len(str)-n-1]):
            return False
        n-=1
        
    return True
```

```python
isPalindrome("abb")
```

```python
res = []
curr_strs = []
s = "aab"

#backtrack to generate all patitions
def backtrack(pos):
    if(pos >= len(s)):
        res.append(curr_strs.copy())
        return
    print(pos)
    for j in range(pos, len(s)):
        if(isPalindrome(s[pos:j+1])):
            curr_strs.append(s[pos:j+1])
            backtrack(j+1)
            curr_strs.pop()
```

```python

backtrack(0)
print(res)

```
