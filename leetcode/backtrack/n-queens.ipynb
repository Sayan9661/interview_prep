```python

def solveNQueens(n: int):
    #sets hold no duplicates and are used to check if there will be conflicts in rows and diagonals
    cols = set()
    posdiag = set()
    negdiag = set()
    #dont need to check rows as we backtrack and go to next row anyways.
    res = []

    board = [["."]*n for i in range(n)]

    def backtrack(row):
        #if all rows complete in backtrack then all n queens placed with no conflict and we have a result
        if(row == n):
            #append the possible board to the result in correct format
            res.append(["".join(r) for r in board])
            return

        for col in range(n):
            #check if there is a conflict with an existing queen if new queen placed in the current row,col
            if (col+row in posdiag) or (row-col in negdiag) or (col in cols):
                continue


            #if no confilct place queen
            cols.add(col)
            posdiag.add(row+col)
            negdiag.add(row-col)
            board[row][col] = "Q"

            #backtrack and go to next row to find out if placement of other queens possible
            backtrack(row+1)

            #remove current placement to try next possible col in current row
            cols.remove(col)
            posdiag.remove(row+col)
            negdiag.remove(row-col)
            board[row][col] = "."

    backtrack(0)
    return res
```
